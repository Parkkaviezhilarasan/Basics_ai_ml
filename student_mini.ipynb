```python
import pandas as pd
import numpy as np
import statistics as stats
data={'Name':['Monika','Subi','Abi','Sri','Parkkavi','Nithya','Kani','Jovita','Shan','Sena'],
      'Math_mark':[98,65,45,78,45,95,np.nan,68,83,79],
      'Science_mark':[95,61,50,82,92,82,72,90,np.nan,70],
      'English_mark':[99,64,np.nan,85,85,79,np.nan,92,85,75],
      'Study_hours':[5,3,2,5,5,np.nan,6,6,5,5],
      'Sleep_hours':[6,8,8,6,7,8,7,7,5,5]}
df=pd.DataFrame(data)
print(df)
```

```python
print("Number of cloumns",len(df.columns))
print("Number of rows",len(df))
print(df.shape)
print("-----------------------------------------------------------")
print(df.describe())
print(df.info())
print("-----------------------------------------------------------")
print(df.count())
print("-----------------------------------------------------------")
print(df.dtypes)
print("-----------------------------------------------------------")
print(df.isnull().sum())
```

```python
mean_Math_marks=df['Math_mark'].mean()
print("Mean of math: ",mean_Math_marks)
median_English_marks=df['English_mark'].median()
print("Median of english: ",median_English_marks)
mode_Science_marks=df['Science_mark'].mode()
print("Mode of science: ",mode_Science_marks)
print("-----------------------------------------------------------")
#another method
mean_mm=np.mean(df['Math_mark'])
median_em=np.median(df['English_mark'])
mode_sm=stats.mode(df['Science_mark'])
print("Mean of math: ",mean_mm)
print("Median of english: ",median_em)
print("Mode of science: ",mode_sm)
print("-----------------------------------------------------------")
#using stats
mean2=stats.mean(df['Math_mark'])
median2=stats.median(df['English_mark'])
mode2=stats.mode(df['Science_mark'])
print("Mean of math: ",mean2)
print("Median of english: ",median2)
print("Mode of science: ",mode2)
```

```python
var_maths=df['Math_mark'].var()
print("Variance of maths",var_maths)
std_maths=df['Math_mark'].std()
print("Standard deviation of maths",std_maths)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(4,4))
sns.histplot(df['Math_mark'], bins=5, kde=True) 
plt.title("Distribution of Math Marks")
plt.xlabel("Math Marks")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,8))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.title("Boxplot for All Numeric Columns")
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include='number').corr()
print(correlation_matrix)
#another method
correlation_matrix = df.drop(columns=['Name']).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
